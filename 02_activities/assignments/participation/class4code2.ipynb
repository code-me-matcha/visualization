```python
# 05 - subplots and combining visualizations
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0,75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110,180,240,99,220]) 
```

```python
# introducing subplots
# in past lessons, we would do the following code to define our figure and its ONE! axes
# fig, ax = plt.subplots(figsize=(5,3))

# now, we want to have 2 plots NEXT TO EACH OTHER
# so we need to define multiple axes and their relative positions
fig, (ax1, ax2) = plt.subplots(ncols = 2,
                                nrows = 1,
                                figsize=(7,3))

 # then, we can use our new figure, with its 2 axes
# and define the types of viz we want to see in each
# graph on left to be scatter; graph on left to be bar
ax1.scatter(x1,y1)
ax2.bar(x2,y2)
fig.show()

# activity: customizing our plots based on everything we remember from previous classes
ax1.annotate("Look!", xy = (15,55), xytext = (30,54),
            arrowprops = dict(arrowstyle = "-["))

font1 = {'family': 'serif', 'color': 'indigo'}
ax1.set_title("Random Scatterplot", fontdict = font1)
ax2.set_title("One Piece Characters", fontdict = font1)
```

```python
# plt.subplot_mosaic(): subplots without a grid arrangement
# we can arrange subplots within our figure usinng plt.subplot_mosaic()

fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                    figsize=(7,4))
# the mosaic is the literal structure of how we want our figure laid out
# so in the above code ax 1 in upper left, ax2 in bottom left
# since ax3 is in both right hand sides (both upper and bottom) then it stretches over that entire side
# so it's basically 3 different types of graphs, all with the SAME AXIS - showing same data

# once we've made our mosaic, we can add data to each of our subplots the same way we did before
# just reference each axes label in our someaxes list
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                    figsize=(7,4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
plt.show()
```

```python
# modifying our figure layout
# adding some very large x axis titles to our previous plot
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                    figsize=(7,4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

someaxes["ax1"].set_xlabel('A Big Label', fontsize = 18)
someaxes["ax2"].set_xlabel('Another Label', fontsize = 18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize=18)
# when we run above code - axis title for ax1 (scatterplot) doesn't appear, and the axis title for line plot (ax3) is cut off!
# so --> we need to use layouts to make sure our subplots fit NEATLY into our figure area
# 2 main kinds of layouts we consider:
#   1) tight layout: adjusts subplots to tick labels, axis labels, and titles don't overlap or leave the figure area
#   2) constrained layout: works similarly, except it also fits things like legends or colourbars

# trying the CONSTRAINED LAYOUT
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                    figsize=(7,4),
                                    layout = "constrained") # added constrained layout here!
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

someaxes["ax1"].set_xlabel('A Big Label', fontsize = 18)
someaxes["ax2"].set_xlabel('Another Label', fontsize = 18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize=18)

# trying the TIGHT LAYOUT
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                    figsize=(7,4),
                                    layout = "tight") # added tight layout here
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

someaxes["ax1"].set_xlabel('A Big Label', fontsize = 18)
someaxes["ax2"].set_xlabel('Another Label', fontsize = 18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize=18)
# overall looks the same as constrained layout since we didn't already have legends or colour bars in the code anyways
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

fig, ax = plt.subplots(figsize=(5,3))
#ax.scatter(x,y) # makes a scatter plot
#ax.bar(x,y) # makes a bar chart
ax.plot(x,y) # makes a line graph
#ax.hist(y) # only need y data for histogram since x counts occurences

# adding plot labels
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')

# modifying labels and titles with different fonts, sizes, colours
font1 = {'family': 'sans-serif', 'color': 'blue', 'size': 20}
font2 = {'family': 'monospace', 'color': 'green', 'size': 14}
```

```python
# adding fontdict arguments to our previous lineplot
# using loc which stands for location to left and right align
# if we don't include a loc argument, it will be centre-aligned
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1, loc = 'left')
ax.set_ylabel('Total growth', fontdict =font2)
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
# modifying the data points in our scatterplot
# we remake our basic scatterplot but this time with color and marker arguments - need to use American spelling
# this basically shows the datapoints
# documentation for more options: https://matplotlib.org/stable/api/markers_api.html
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,
            y,
            marker = '*', # shows each datapoint as a star, can use > to show as arrows
            color = 'indigo') # in indigo color

```

```python
# modifying the data points to add lines
# want a line plot so we use ax.plot
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,
        y,
        marker = '*',
        color = 'indigo',
        linestyle = '--', # dashed line connecting our markers
        linewidth = 2)
fig.show() # this becomes useful when we're working with multiples axes - so it explicitly tells it to print out the whole figure and not only the axes we're working with
# only works with ax.plot - we need to tell python to make a line chart!
```

```python
# changing the colours
# if we want to use a specific colour, can use hex codes
# further customizations
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,
        y,
        marker = '*',
        markersize = 12,
        color = '#7425b9',
        linestyle = '--',
        linewidth = 2,
        markeredgecolor = '#fa9359', # this is the outline of the stars
        markerfacecolor = '#000000') # this is the fill colour
```

```python
# adding grid lines to our axis object using .grid()
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,
        y,
        marker = '*',
        markersize = 12,
        color = '#7425b9',
        linestyle = '--',
        linewidth = 2,
        markeredgecolor = '#fa9359',
        markerfacecolor = '#000000')
ax.grid(axis = 'y',
        linestyle = '--',
        linewidth = 3,
        color = '#fa9359')

# cheat sheets
 #https://s3.amazonaws.com/assets.datacamp.com/blog_assets/Python_Matplotlib_Cheat_Sheet.pdf
 #https://github.com/matplotlib/cheatsheets (w
```

```python
# activity: copying the code for a graph from the python graph gallery
# source: https://python-graph-gallery.com/parallel-coordinate-plot-plotly/

# import libraries
import plotly.express as px

# load the iris dataset provided by the library
df = px.data.iris()

# create the chart
fig = px.parallel_coordinates(
    df,
    color = "species_id",
    labels = {"species_id": "Species", "sepal_width": "Sepal Width", "sepal_length": "Sepal Length", "petal_width": "Petal Width", "petal_length": "Petal Length"},
    color_continuous_scale = px.colors.diverging.Tealrose,
    color_continuous_midpoint = 2
)

# hide the color scale since it is useless in this case
fig.update_layout(coloraxis_showscale=False)

# show the plot
fig.show()
```
